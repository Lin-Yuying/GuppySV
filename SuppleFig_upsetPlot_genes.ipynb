```python
import upsetplot
import pandas as pd
from upsetplot import UpSet, from_memberships
import matplotlib.pyplot as plt
from upsetplot import plot
```

```python

memberships = [
    ('Aripo',),
    ('Quare',),
    ('Yarra',),
    ('Aripo', 'Quare'),
    ('Aripo', 'Yarra'),
    ('Quare', 'Yarra'),
    ('Aripo', 'Quare', 'Yarra')
]

counts = [345, 365, 550, 164, 265, 254, 138]

# Create the Series with MultiIndex boolean index
data = from_memberships(memberships, data=counts)


fig = plt.figure(dpi=500, figsize=(10, 6))

# Panel A: UpSet
upset = UpSet(data, sort_by='degree', show_counts='%d')
upset_plot = upset.plot(fig=fig)

upset_plot['intersections'].set_ylabel("No. of genes")




plt.show()
```
