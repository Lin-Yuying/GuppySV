```python
import rpy2
from rpy2 import robjects
```

```python
from rpy2.robjects.packages import importr
```

```python
lostruct = importr("lostruct")
```

```python
bcf_file = "/Users/rainlam/Project/3.SV/Aripo.LG4.vcf.gz"
window_size = 100
```

```python
sites = lostruct.vcf_positions(bcf_file)
```

```python
win_fn_snp = lostruct.vcf_windower(bcf_file, size=window_size, type="snp", sites=sites)
```
