```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt, mpld3
```

```python
def findPop(data='/Users/rainlam/Project/3.SV/PCA/data.pca.eigenvec'):
    sex = []
    pop = []
    df = pd.read_csv(data, sep='\t', header=0)
    for i in list(df['IID']):
        i = i.upper()
        sex.append(i[4])
        pop.append(i[:3])
    df['POPULATION'] = pop
    df['SEX'] = sex
    return df

df = findPop()
```
